```python
import networkx as nx
import matplotlib.pylab as plt
```

```python
def chain(G):
    node = max(G.nodes)
    parents = [e[0] for e in G.in_edges(node)]
    children = [e[1] for e in G.out_edges(node)]
    G.remove_node(node)
    G.add_nodes_from([node,node+1])
    G.add_edges_from([(node,node+1)])
    G.add_edges_from([(i,node) for i in parents])
    G.add_edges_from([(node+1,i) for i in children])
    return G
```

```python
def partition(G):
    node = max(G.nodes)
    parents = [e[0] for e in G.in_edges(node)]
    children = [e[1] for e in G.out_edges(node)]
    G.remove_node(node)
    G.add_nodes_from([node,node+1])
    G.add_edges_from([(i,node) for i in parents])
    G.add_edges_from([(node+1,i) for i in children])
    return G
```

```python
def order(G):
    node = max(G.nodes)
    parents = [e[0] for e in G.in_edges(node)]
    children = [e[1] for e in G.out_edges(node)]
    G.remove_node(node)
    G.add_nodes_from([node,node+1])
    G.add_edges_from([(node,node+1)])
    G.add_edges_from([(i,node) for i in parents])
    G.add_edges_from([(i,node+1) for i in parents])
    G.add_edges_from([(node,i) for i in children])
    G.add_edges_from([(node+1,i) for i in children])
    return G
```

```python
def hierarchy(G,node=None):
    if node == None:
        node = max(G.nodes)
    else:
        assert node in G
    new_node = max(G.nodes)+1
    G.add_nodes_from([new_node])
    G.add_edges_from([(node,new_node)])
    return G
```

```python
def tree(G):
    node = max(G.nodes)
    new_node_1 = max(G.nodes)+1
    new_node_2 = max(G.nodes)+2
    G.add_nodes_from([new_node_1,new_node_2])
    G.add_edges_from([(node,new_node_1),(node,new_node_2)])
    return G
```

```python
def repeat(func, n, x):
    for i in range(n):
        x = func(x)
    return x
```
