```python
import numpy as np
import matplotlib.pylab as plt
```

```python
%matplotlib inline
```

```python
N = int(1e7)
```

```python
flips = np.random.choice(2, size = N)*2-1
```

```python
walk = np.cumsum(flips)
```

```python
# E[|n_heads - n_tails|] = sqrt(2*N/pi)
```

```python
plt.plot(np.arange(N),walk);
plt.plot(np.arange(N),np.sqrt(2*np.arange(N)/np.pi));
```

```python
plt.plot(np.arange(N),np.cumsum(flips)/np.arange(N));
plt.ylim((-0.005,0.005))
```
