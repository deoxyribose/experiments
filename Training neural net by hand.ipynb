```python
import numpy as np
import matplotlib.pylab as plt
```

```python
x1r = np.arange(0,1,0.01)
x2r = np.arange(0,1,0.01)
```

```python
x1 = np.random.rand(100)
x2 = np.random.rand(100)
```

```python
def classify(x1,x2):
    if x2 <= 0.75 and x2 <= 1-x1 and x1 <= 0.75:
        return 1
    else:
        return 0
```

```python
labels = [classify(x,y) for x,y in zip(x1,x2)]
```

```python
plt.plot(x1,np.ones_like(x1)*0.75)
plt.plot(x1,1-x1)
plt.plot(np.ones_like(x2)*0.75,x2)
plt.scatter(x1,x2,c=labels)
plt.xlim(0,1)
plt.ylim(0,1)
```

```python
def a(w,x,bias):
    return (np.dot(w,x.T)+bias >= 0).astype(np.int32)
```

```python
plt.plot(x2r,a(np.array([0.,1.]),np.array([x1r,x2r]).T,-0.75))
```

```python
plt.plot(x1r,a(np.array([-1.,0.]),np.array([x1r,x2r]).T,0.75))
```

```python
# x2 <= 1-x1
# <=>
# -1 <= -x2-x1
```

```python
plt.plot(x1,1-x1)
plt.scatter(x1,x2,c=a(np.array([-1.,-1.]),np.array([x1,x2]).T,1.))
plt.xlim(0,1)
plt.ylim(0,1)
```

```python
# known weights
w1 = 0
w2 = -1
w8 = 0.75
```

```python
plt.scatter(x1,x2,c=0.5*h1-0.5*h2-0.5*h3+1.>0.)
plt.plot(x1,np.ones_like(x1)*0.75)
plt.plot(x1,1-x1)
plt.plot(np.ones_like(x2)*0.75,x2)
```
