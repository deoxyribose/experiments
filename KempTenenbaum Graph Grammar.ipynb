```python
def chain(G):
    node = max(G.nodes)
    parents = [e[0] for e in G.in_edges(node)]
    children = [e[1] for e in G.out_edges(node)]
    G.remove_node(node)
    G.add_nodes_from([node,node+1])
    G.add_edges_from([(node,node+1)])
    G.add_edges_from([(i,node) for i in parents])
    G.add_edges_from([(node+1,i) for i in children])
    return G
```

```python
def partition(G):
    node = max(G.nodes)
    parents = [e[0] for e in G.in_edges(node)]
    children = [e[1] for e in G.out_edges(node)]
    G.remove_node(node)
    G.add_nodes_from([node,node+1])
    G.add_edges_from([(i,node) for i in parents])
    G.add_edges_from([(node+1,i) for i in children])
    return G
```

```python
def order(G):
    node = max(G.nodes)
    parents = [e[0] for e in G.in_edges(node)]
    children = [e[1] for e in G.out_edges(node)]
    G.remove_node(node)
    G.add_nodes_from([node,node+1])
    G.add_edges_from([(node,node+1)])
    G.add_edges_from([(i,node) for i in parents])
    G.add_edges_from([(i,node+1) for i in parents])
    G.add_edges_from([(node,i) for i in children])
    G.add_edges_from([(node+1,i) for i in children])
    return G
```

```python
def hierarchy(G,node=None):
    if node == None:
        node = max(G.nodes)
    parents = [e[0] for e in G.in_edges(node)]
    children = [e[1] for e in G.out_edges(node)]
    G.remove_node(node)
    G.add_nodes_from([node,node+1])
    G.add_edges_from([(node,node+1)])
    G.add_edges_from([(i,node) for i in parents])
    G.add_edges_from([(node,i) for i in children])
    return G
```

```python
nx.draw(hierarchy(G,2), with_labels=True)
```
